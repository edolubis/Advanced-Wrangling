```python
import pandas as pd
import numpy as np
```

## 1. Loading Sales Data

```python
## Loading Data & Directory sesuaikan dengan komputer masing-masing

df_product_dataset = pd.read_csv('Documents/LAB-BI2/Data Sales/olist_products_dataset.csv')
df_customer_dataset = pd.read_csv('Documents/LAB-BI2/Data Sales/olist_customers_dataset.csv')
df_order_item_dataset = pd.read_csv('Documents/LAB-BI2/Data Sales/olist_order_items_dataset.csv')
df_order_review_dataset = pd.read_csv('Documents/LAB-BI2/Data Sales/olist_order_reviews_dataset.csv')
df_orders_dataset = pd.read_csv('Documents/LAB-BI2/Data Sales/olist_orders_dataset.csv')
df_sellers_dataset = pd.read_csv('Documents/LAB-BI2/Data Sales/olist_sellers_dataset.csv')
```

## 2. Merging Every Sales Data

```python
dfmerged1 = pd.merge(df_order_review_dataset,df_orders_dataset, how='inner')
dfmerged2 = pd.merge(df_order_item_dataset,dfmerged1,how='inner')
dfmerged3 = pd.merge(dfmerged2,df_product_dataset, how='inner')
dfmerged4 = pd.merge(dfmerged3,df_sellers_dataset, how='inner')
dfmerged5 = pd.merge(dfmerged4,df_customer_dataset, how='inner')
dfmerged5
```

## 3. Check and Drop Null Value

```python
dfmerged5.isna().sum()
```

```python
total_na = dfmerged5.isnull().sum().sort_values(ascending=False)
percent = ((dfmerged5.isnull().sum()/dfmerged5.isnull().count())*100).sort_values(ascending=False)
na_data = pd.concat([total_na,percent], axis=1,keys=['Total_NA','Percent_%'])

na_data
```

### Karena Column "review_comment_title" dan "review_comment_message" memiliki file N/A lebih dari 50%, maka tidak didrop melainkan diisi sesuai rating yang diberikan pembeli

# Fill N/A pada column 'review_comment_title' dan 'review_comment_message' karena file N/A >50%

```python
## Isi komentar sesuai dengan rating score yang diberikan Pembeli

dfmerged5.loc[(dfmerged5.review_score == 5),'review_comment_title'] = dfmerged5['review_comment_title'].fillna('Maravilhoso')
dfmerged5.loc[(dfmerged5.review_score == 4),'review_comment_title'] = dfmerged5['review_comment_title'].fillna('Muito bom')
dfmerged5.loc[(dfmerged5.review_score == 3),'review_comment_title'] = dfmerged5['review_comment_title'].fillna('poderia ser melhor')
dfmerged5.loc[(dfmerged5.review_score == 2),'review_comment_title'] = dfmerged5['review_comment_title'].fillna('estou desapontado')
dfmerged5.loc[(dfmerged5.review_score == 1),'review_comment_title'] = dfmerged5['review_comment_title'].fillna('Muito desapontado')


dfmerged5.loc[(dfmerged5.review_score == 5),'review_comment_message'] = dfmerged5['review_comment_message'].fillna('Perfeito')
dfmerged5.loc[(dfmerged5.review_score == 4),'review_comment_message'] = dfmerged5['review_comment_message'].fillna('Muito bom')
dfmerged5.loc[(dfmerged5.review_score == 3),'review_comment_message'] = dfmerged5['review_comment_message'].fillna('poderia ser melhor')
dfmerged5.loc[(dfmerged5.review_score == 2),'review_comment_message'] = dfmerged5['review_comment_message'].fillna('estou desapontado')
dfmerged5.loc[(dfmerged5.review_score == 1),'review_comment_message'] = dfmerged5['review_comment_message'].fillna('Muito desapontado')
```

## Cek %  Data N/A

```python
total_na = dfmerged5.isnull().sum().sort_values(ascending=False)
percent = ((dfmerged5.isnull().sum()/dfmerged5.isnull().count())*100).sort_values(ascending=False)
na_data = pd.concat([total_na,percent], axis=1,keys=['Total_NA','Percent_%'])

na_data
```
